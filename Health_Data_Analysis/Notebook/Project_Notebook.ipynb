### Visualizing Correlations Between Variables:
Use scatter plots or correlation matrices to explore relationships between continuous variables.

```python
#09. creating a df contion Income and BMI columns.
continuous_vars = ['Income', 'BMI']
df_continuous = df[continuous_vars]

# Calculate the correlation matrix
correlation_matrix = df_continuous.corr()
correlation_matrix
```

#### A correlation matrix shows how different variables relate to each other. 
#### The values in the matrix, called correla􀆟on coefficients, range from -1 to +1.
##### Key Components:
The Numbers: The values indicate the strength and direc􀆟on of the rela􀆟onship.
- +1.00: A perfect positive correlation. This means when one variable increases, the other
increases in perfect propor􀆟on.
- -1.00: A perfect negative correlation. This means when one variable increases, the other
decreases in perfect proportion.
- 0: No linear relationship.
- Values closer to +1 or -1 show a stronger rela􀆟onship, while values closer to 0 show a weaker rela􀆟onship.

```python
#plotting
plt.figure(figsize=(5, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

#### Here, in correla􀆟on heatmap of Income and BMI shows that:
- Diagonal Cells: The cells where a variable is correlated with itself (e.g., "Income" with "Income") always have a value of 1.00.
- This is expected, as a variable is perfectly correlated with itself.\
Off-Diagonal Cells: The cells comparing two different variables (e.g., "Income" and "BMI") contain the meaningful informa􀆟on.
- The value is -0.10. This is a weak nega􀆟ve correla􀆟on between income and BMI.\
In context, this means that as an individual's income increases, their BMI tends to slightly decrease, but the rela􀆟onship is not strong.

### Compare Heart Disease Across Demographic Groups:
Analyze the prevalence of heart disease across different demographic groups (e.g., age, gender, education) using bar charts or pivot tables.

```python
# 13. Create a pivot table to find heart disease prevalence by age
age_prevalence = df.pivot_table(
    index='Age',
    values='HeartDiseaseorAttack',
    aggfunc='mean'
) * 100
age_prevalence
```

```python
#plotting
age_prevalence.plot(kind='bar', color='skyblue', rot = 0)
plt.xlabel('Age Distrubution')
plt.ylabel('Percentage of HeartDiseaseorAttack')
plt.show()
```


**Bar graph:** This show the percentage of HeartDiseaseorA􀆩ack with different age groups:\
First calculates the prevalence of heart disease for each age group and stores it in a new table
called age_prevalence.\
- By using **df.pivot_table()** func􀆟on , it will contain two columns Age and HeartdiseaseorA􀆩ack .
- Use **aggfunc='mean':** It calculates the mean (average) of the HeartDiseaseorAttack values
for each age group.
- *100: This mul􀆟plies the propor􀆟on by 100 to convert it into a percentage, making the result
easier to interpret.\
here we find the **clear upward trend:** as the age group number increases, the percentage of individuals with heart disease consistently and **significantly rises**.\
This demonstrates a **strong positive correlation** between age and the prevalence of heart disease.
